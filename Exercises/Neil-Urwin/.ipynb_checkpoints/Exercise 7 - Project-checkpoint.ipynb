```python
train = pd.read_csv('data/train.csv', parse_dates=['DateTime'])
```

# Exploratory Data Analysis

```python
def convert_age(col):
    try:
        age = str(col).split(' ')
        if (age[1] == 'year') | (age[1] == 'years'):
            return int(age[0])
        else:
            return 0
    except:
        pass
train['AgeInYearsUponOutcome'] = train['AgeuponOutcome'].apply(convert_age)
```
