# Solutions to matplotlib and seaborn exercises

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Question 1
create a bar chart of Pokemon species introduced by generation.

```python
# data setup
pokemon = pd.read_csv('../data/pokemon.csv')
fuel_econ = pd.read_csv('../data/fuel_econ.csv')
```

```python
# pokemon
sns.countplot(data=pokemon, x='generation_id') # or sns.countplot(x=pokemon['generation_id'])
sns.despine()
```

## Question 2

create a histogram of Pokemon special defense values.

```python
sns.distplot(pokemon['defense'])
sns.despine()
```

## Question 3

create a histogram of Pokemon heights and weights.

There's a very long tail of Pokemon heights. Here, I've focused in on Pokemon of height 6 meters or less, so that I can use a smaller bin size to get a more detailed look at the main data distribution.

```python
n, bins, patches = plt.hist(pokemon['height'], range=(0,6), bins=25)
plt.title('Pokemon Height (6m or less)')
plt.xlabel('Height (m)')
plt.show()
```

```python
n, bins, patches = plt.hist(pokemon['weight'])
plt.title('Pokemon Weight')
plt.xlabel('Weight (kg)')
plt.show()
```

## Question 4
create a scatterplot of city vs. highway fuel mileage.

```python
#fuel_econ
axes = sns.scatterplot(x=fuel_econ['city'], y=fuel_econ['highway'])
axes.set(xlabel='City Fuel Eff. (mpg)', ylabel='Highway Fuel Eff. (mpg)')
sns.despine()
```
