## Step 1. Import the necessary libraries

```python
# Import necessary libraries
import numpy as np 
import pandas as pd

# Fancy display object
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

## Step 2. Create a data dictionary

```python
# Create data dictonary of each Pokémon, their stats and the next stage of their evolution
evo = pd.Series({'0': 'Ivysaur', '1': 'Charmeleon', '2': 'Wartotle', '3': 'Metapod'})
hp = pd.Series({'0': 45, '1': 39, '2': 44, '3': 45})
nam = pd.Series({'0': 'Bulbasaur', '1': 'Charmander', '2': 'Squirtle', '3': 'Caterpie'})
pkdex = pd.Series({'0': 'yes', '1': 'no', '2': 'yes', '3': 'no'})
typ = pd.Series({'0': 'grass', '1': 'fire', '2': 'water', '3': 'bug'})
```

## Step 3. Assign it to a variable

```python
# Create data frame and assign to a variable
pokemon = pd.DataFrame({'evolution': evo, 'hp': hp, 'name': nam, 'pokedex': pkdex, 'type': typ})
pokemon
```

## Step 4. Place the order of of the columns as name, type, hp, evolution, pokedex

```python
# Reorder columns
pokemon = pokemon.reindex(['name','type','hp','evolution','pokedex'], axis=1)
pokemon
```

## Step 5. Add another column called Place and insert what you have in mind

```python
# Create the location of each Pokémon and merge it with the data frame
place = pd.Series({'0': 'Pallet Town', '1': 'Cinnabar Island', '2': 'Cerulean City', '3': 'Viridian Forest'})
pokeplaces = pd.DataFrame({'name': nam, 'place': place})
display('pokemon', 'pokeplaces', 'pd.merge(pokemon, pokeplaces)')
```

```python
# Alternative means of merging the data dictonary with the data frame
pokemon['place'] = pd.Series({'0': 'Pallet Town', '1': 'Cinnabar Island', '2': 'Cerulean City', '3': 'Viridian Forest'})
pokemon
```
